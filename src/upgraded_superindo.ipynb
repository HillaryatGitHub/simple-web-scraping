```python
Array = []
for elemen in array:
    x = elemen.strip("Tel ")
    Array.append(x)
```

```python
hitung_slash = 0
total_slash = 0
for elemen in Array:
    hitung_slash = elemen.count('/')
    total_slash = total_slash + hitung_slash
print("Total Nomor Telp Tambahan: " + str(total_slash))
```

```python
def slash_each_element(Array):
    jumlah_slash = []
    
    for element in Array:
        slash = element.count("/")
        jumlah_slash.append(slash)
    return jumlah_slash

deret_slash = slash_each_element(Array)
print(deret_slash)
print("Maximum Slash", max(deret_slash))
# Pada 1 cabang, terdapat maksimal 3 nomor telepon pada data
```

```python
first_array = []
second_array = []
third_array = []
    
for element in Array:
    slash = element.count("/")
    if slash == 0:
        first_array.append(element)
        second_array.append("")
        third_array.append("")
    elif slash == 1:
        pisah = element.split("/")
        first_array.append(pisah[0])
        second_array.append(pisah[1])
        third_array.append("")
    elif slash == 2:
        pisah = element.split("/")
        first_array.append(pisah[0])
        second_array.append(pisah[1])
        third_array.append(pisah[2])

print("Nomor Pertama", first_array)
print("Nomor Kedua", second_array)
print("Nomor Ketiga", third_array)
```

```python
#element = "30 4 - 450"
element = "304 - 4550"
elements = element.split("-")

for i in range(len(elements)):
    numbers = elements[i].split()
    combined_number = "".join(numbers)
    elements[i] = combined_number

modified_element = " - ".join(elements)

print(modified_element)
```
